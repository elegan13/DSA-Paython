## 4. Which code is appropriate to drop duplicate values in 'App' column and remove its rows in Python?  Assume dataset was loaded as a pandas data frame variable: google_play_apps.

### The Answer:

```python
#Step1 - Check if there is duplication or not
#Sorting by App Names
#Count by App Names
count_app_names = google_play_apps.groupby(['App']).size().reset_index(name='Counts')
sort_count_app_names = count_app_names.sort_values(by=['Counts','App'],ascending=False)
sort_count_app_names.head(20)
```

```python
#Step 2 - Remove duplicates with inplace dupicate remove
google_play_apps.drop_duplicates(subset='App', inplace=True)
count_app_names = google_play_apps.groupby(['App']).size().reset_index(name='Counts')
sort_count_app_names = count_app_names.sort_values(by=['Counts','App'],ascending=True)
sort_count_app_names.head(20)
```

```python
#Step 3 - Check rows numbers
google_play_apps.head(10)
print("rows:",len(google_play_apps))
```

#### #So the code for to drop duplicate values in 'App' column and remove its rows in Python is "google_play_apps.drop_duplicates(subset='App', inplace=True)"

## 9.Arrange top COMMUNICATION category apps from the highest number of reviews to the lowest!
